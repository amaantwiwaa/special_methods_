## 1. Constructor: __init__
Purpose: Initializes a newly created object's attributes.
Parameters: self (the instance), followed by custom arguments.

```python
class Book:
    def __init__(self, title):
        self.title = title

b = Book("Python 101")
```

## 2. String Representation: __str__ and __repr__
Purpose: __str__ defines a user-friendly string view. __repr__ defines an unambiguous developer-facing string view.Parameters: self only.

```python
class Book:
    def __str__(self):
        return "A great book"
    def __repr__(self):
        return "Book()"

b = Book()
print(str(b))   # A great book
print(repr(b))  # Book()
```

## 3. Length: __len__
Purpose: Allows an object to return its size when passed to the global len() function.Parameters: self only.

```python
class Shelf:
    def __init__(self):
        self.items = ["Book1", "Book2"]
    def __len__(self):
        return len(self.items)

s = Shelf()
print(len(s))  # 2
```

## 4. Equality: __eq__
Purpose: Defines how objects are compared using the == operator.Parameters: self, other (the object being compared against).

```python
class Item:
    def __init__(self, price):
        self.price = price
    def __eq__(self, other):
        return self.price == other.price

print(Item(10) == Item(10))  # True
```

## 5. Operator Overloading: __add__, __sub__, __mul__
Purpose: Intercepts arithmetic operators (+, -, *) when applied to custom objects.Parameters: self, other (the right-hand side operand).

```python
class Point:
    def __init__(self, x):
        self.x = x
    def __add__(self, other):
        return Point(self.x + other.x)

p3 = Point(5) + Point(10)
```

## 6. Item Access: __getitem__
Purpose: Enables indexing or bracket notation (e.g., obj[key]) to fetch data.Parameters: self, key (index or dictionary key).

```python
class CustomList:
    def __init__(self):
        self.data = [10, 20, 30]
    def __getitem__(self, index):
        return self.data[index]

c = CustomList()
print(c[1])  # 20
```

## 7. Item Assignment: __setitem__

Purpose: Enables bracket assignment notation (e.g., obj[key] = value) to modify data.Parameters: self, key (index), value (the item to assign).

```python
class CustomList:
    def __init__(self):
        self.data = [0, 0]
    def __setitem__(self, index, value):
        self.data[index] = value

c = CustomList()
c[0] = 99
```

## 8. Membership Testing: __contains__
Purpose: Overrides the in operator behavior to check if an element exists inside the object.Parameters: self, item (the value to check).

```python
class Team:
    def __init__(self):
        self.members = ["Alice", "Bob"]
    def __contains__(self, item):
        return item in self.members

t = Team()
print("Alice" in t)  # True
```

## 9. Callable Objects: __call__
Purpose: Allows an instance of a class to be called exactly like a standard function (e.g., obj()).Parameters: self, followed by any optional positional or keyword arguments.

```python
class Greeter:
    def __call__(self, name):
        return f"Hello {name}"

greet = Greeter()
print(greet("John"))  # Hello John
```

## 10. Iteration Support: __iter__ and __next__
Purpose: __iter__ returns an iterator object. __next__ retrieves the subsequent item during loops.Parameters: self only for both methods.

```python
class Counter:
    def __init__(self):
        self.num = 0
    def __iter__(self):
        return self
    def __next__(self):
        self.num += 1
        if self.num > 2:
            raise StopIteration
        return self.num

for x in Counter():
    print(x)  # Prints 1, then 2
```

## 11. Boolean Evaluation: __bool__
Purpose: Defines whether an object evaluates to True or False in conditional statements.Parameters: self only.

```python
class Account:
    def __init__(self, balance):
        self.balance = balance
    def __bool__(self):
        return self.balance > 0

acc = Account(100)
if acc:
    print("Active account")
```

## 12. Destructor: __del__
Purpose: Triggered automatically when an object is about to be destroyed or garbage-collected.Parameters: self only.

```python
class Resource:
    def __del__(self):
        print("Cleaned up resource")

r = Resource()
del r  # Outputs: Cleaned up resource
```
